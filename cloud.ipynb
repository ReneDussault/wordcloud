```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# lists of words

glor = ["honest", "level headed", "clear thinking", "rational", "patient", "humble", "love for learning", "handsome"]

phil = ["meticulous", "patient", "calm", "loves to share knowledge", "creative"]

karim = ["perseverance", "good communicator", "team spirit", "rigorous", "versatility", "meticulous", "love for learning", "adaptability"]

charly = ["rigorous", "love for learning", "loves to share knowledge", "resourceful"]

shannon = ["honest", "resourceful", "kind", "patient", "funny", "thoughtful", "curious", "competent"]
```

```python
all_words = glor + phil + karim + charly + shannon
```

```python
# let's replace "loves to share knowledge" with mentor (educator could also work) and "love for learning" with learner (as in life long learner)

replace_dict = {
    "loves to share knowledge": "mentor",
    "love for learning": "learner"
}


all_words = [replace_dict.get(word, word) for word in all_words]
```

```python
# adding an underscore on some of the strings 

und_words = [word.lower().replace(" ", "_") for word in all_words]
```

```python
und_words
```

```python
processed_text = ' '.join(und_words)
```

```python
wordcloud = WordCloud(width=800, height=600, background_color='black').generate(processed_text)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
